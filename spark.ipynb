```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('spark').master('local[*]').getOrCreate()
```

```python
parquet_file1 = "part-00000-4fe26c42-bc73-4826-add1-a7156f130736-c000.snappy.parquet"
data1 = spark.read.parquet(parquet_file1)

parquet_file2 = 'part-00000-31b4f4cb-7770-4a5f-9e1e-c9118810d7f0-c000.snappy.parquet'
data2 = spark.read.parquet(parquet_file2)
```

```python
data = data1.union(data2).distinct()
```

```python
data.count()
```

```python
data.select('*').show(5)
```

```python
from pyspark.sql import functions as f
from pyspark.sql.functions import lower, col
```

```python
data.select(
    '*',
    f.when(lower(col('purpose')).like('%кредит%') | lower(col('method_receipt')).like('%кредит%') | lower(col('bank_name')).like('%кредит%'), 1).otherwise(0).alias('Есть слово кредит')
).show(10, truncate=False)
```

```python
data_with_credit = data.filter((lower(col("method_receipt")).like("%кредит%"))|(lower(col("bank_name")).like("%кредит%"))|(lower(col("purpose")).like("%кредит%")))
data_with_credit.show(truncate=False)
```

```python
data.filter((lower(col("method_receipt")).like("%кредит%"))|
            (lower(col("bank_name")).like("%кредит%"))|
            (lower(col("purpose")).like("%кредит%")))
```

```python
data_with_credit.count()
```
